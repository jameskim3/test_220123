```python
import numpy as np
import sklearn.metrics as skmetrics
```

```python
labels=np.random.randint(0,2,500)
preds=np.random.random(500)
```

```python
auc_score=skmetrics.roc_auc_score(labels,preds)
auc_score
```

```python
acc=skmetrics.accuracy_score(labels,preds>0.5)
acc
```

```python
fpr,tpr,thresholds=skmetrics.roc_curve(labels,preds)
```

```python
import seaborn as sns
print()
sns.lineplot(x=fpr,y=tpr)
```

```python
norm1=np.random.normal(3,1,size=1000)
norm2=np.random.normal(8,1,size=1000)
```

```python
import matplotlib.pyplot as plt
f,ax=plt.subplots(1,2,figsize=(14,4))
sns.distplot(norm1,ax=ax[0],bins=100,color="g")
sns.distplot(norm2,ax=ax[0],bins=100,color="b")
sns.distplot(norm2,ax=ax[1],bins=100)
```

```python
norm1=np.array(norm1).reshape(-1,1)
norm2=np.array(norm2).reshape(-1,1)
norm3=np.vstack((norm1,norm2))
```

```python
import pandas as pd
df=pd.DataFrame({"n":norm3.flatten()})
df["labels"]=-1
df.loc[1000:,"labels"]=1
df.loc[0:1000,"labels"]=0
```

```python
labels=df.labels.values
preds=df.n.values
fpr,tpr,thresholds=skmetrics.roc_curve(labels,preds)
```

```python
sns.lineplot(x=fpr,y=tpr)
print(f"roc_auc_score:{skmetrics.roc_auc_score(labels,preds)}")
```
