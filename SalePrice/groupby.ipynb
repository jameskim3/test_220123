```python
def category_age(x):
    if x<10:
        return "A"
    if x<12:
        return "B"
    if x<14:
        return "C"
    if x<16:
        return "D"
    if x<18:
        return "E"
    if x<20:
        return "F"
    
    return "G"
```

```python
train['Substrate_new']= train['col1'].apply(category_age)
```

```python
train["Substrate_new"]=train["Substrate"]+"_"+train["Substrate_new"]
```

```python
train.groupby("Substrate_new").mean()
```
